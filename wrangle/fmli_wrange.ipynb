```python
import pandas as pd
```

```python
filename = 'https://raw.githubusercontent.com/timstacey/cex-data-process/master/fmli131x.csv'
df = pd.read_csv(filename)
```

```python
df.describe()
```

```python
# selected_columns = pd.read_clipboard()
selected_columns_f = 'selected_columns.csv'
# selected_columns.to_csv('selected_columns.csv')
selected_columns_df = pd.read_csv(selected_columns_f)
selected_columns_df.columns
```

```python
selected_columns = selected_columns_df['Code'].values
```

```python
selected_columns
```

```python
selected_df = df[selected_columns]
```

```python
selected_df.describe()
```

```python
from collections import defaultdict

categories = defaultdict(set)

for _, column, category in selected_columns_df.values:
    categories[category].add(column)
   
cluster_df = df[list(categories['Stats'])].reindex(index=None)
  
    
for category, column in categories.items():
    if category != 'Stats':
        cluster_df[category] = df[list(column)].sum(axis=1).copy()
```

```python
cluster_file = 'cluster_nonfillna.csv'

cluster_df.to_csv(cluster_file)
```
