```python
from collections import defaultdict

categories = defaultdict(set)

for _, column, category in selected_columns_df.values:
    categories[category].add(column)
   
cluster_df = df[list(categories['Stats'])].reindex(index=None)
  
    
for category, column in categories.items():
    if category != 'Stats':
        cluster_df[category] = df[list(column)].sum(axis=1).copy()
```

```python
cluster_file = 'cluster_nonfillna.csv'

cluster_df.to_csv(cluster_file)
```

```python
household_earners = {   1:'Reference person only', 2:'Reference person and spouse',
                        3:'Reference person, spouse and others',
                        4:'Reference person and others', 
                        5:'Spouse only', 
                        6:'Spouse and others', 
                        7:'Others only', 
                        8:'No earners' }

housing_tenure = {1:'Owned with mortgage',  ### in later more advanced use this
                  2:'Owned without mortgage', 
                  3:'Owned mortgage not reported', 
                  4:'Rented', 
                  5:'Occupied without payment of cash rent', 
                  6:'Student housing' }

housing_tenure_mvp = {1:'own', 
                  2:'own', 
                  3:'own', 
                  4:'rent', 
                  5:'rent', 
                  6:'rent' }
```

```python
housing_tenure # Doing this before it makes the columns more interpretatble
```

```python
df['housing_tenure'] = df['CUTENURE'].map(housing_tenure_mvp)
```

```python
features = ['FSALARYX', 'FAM_SIZE', 'housing_tenure', 'STATE' ]
df_features = df[features]
rent_own = pd.get_dummies(df.housing_tenure)
```

```python
mvp_cluster = pd.merge(df[features], rent_own, left_index=True, right_index=True, how='outer')
columns = ['salary', 'family_size', 'housing_tenure', 'state', 'own', 'rent']
mvp_cluster.columns = columns
```

```python
mvp_cluster_csv = 'mvp_cluster_csv_2015-0801.csv'
mvp_cluster.to_csv(mvp_cluster_csv, index=False)
```
