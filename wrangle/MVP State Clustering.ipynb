```python
from sklearn.cluster import KMeans # K means model
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

%pylab inline
```

```python
mvp_cluster_csv = 'mvp_cluster_csv_2015-0801.csv'
```

```python
data = pd.read_csv(mvp_cluster_csv)
```

```python
salary_limit = 70000
data = data[data.salary < salary_limit] #### dropping all salaries over 70k
```

```python
# data.describe()
```

```python
data.columns
```

```python
features = ['salary', 'own', 'rent'] # 'family_size', ]
```

```python
X = data[features]
```

```python
# X.head()
```

#SCALE DATA

```python
# Standardize our data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
data_scaled = scaler.fit_transform(X)
```

```python
# data_scaled
```

```python
np.random.seed(0)
```

# Train Model

```python
est = KMeans(n_clusters=5, init='random') # Instatiate estimator
est.fit(data_scaled) # Fit your data
y_kmeans = est.predict(data_scaled) # Make cluster "predictions"
```
