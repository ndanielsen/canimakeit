```python
from sklearn.cluster import KMeans # K means model
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

%pylab inline
```

#SCALE DATA

```python
# Standardize our data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
data_scaled = scaler.fit_transform(X)
```

# Predict State Cluster by User Input

```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.cross_validation import train_test_split
from sklearn import metrics
```

```python
state_dummy_vars = pd.get_dummies(data.state)
```

```python
mvp_state_predict = pd.merge(data, state_dummy_vars, left_index=True, right_index=True, how='outer')
```

```python
mvp_state_predict.head()
```

```python
X = mvp_state_predict._get_numeric_data()
```

```python
del X['cluster']
```

```python
y = mvp_state_predict['cluster']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 2)

# Fit model
logreg = LogisticRegression()
logreg.fit(X_train, y_train)
y_pred = logreg.predict(X_test) # Predict

# Access accuracy
print metrics.accuracy_score(y_test, y_pred)
```

```python
def user_request(user):
    salary = int(user['salary']) * 1000
    family_size = int(user["household-size"])
    if user['is-rent'] == 'false':
        own = 1
        rent = 0
    else:
        rent = 1
        own = 0
    return [salary, family_size, own, rent]
    
```

```python
def mvp_state_clusters(user):
    request = user_request(user)
    state_matrix = {}
    states = [u'AK', u'AL', u'AZ', u'CA', u'CO', u'CT', u'DC', u'DE', u'FL', u'GA', u'HI', u'ID', u'IL', u'IN', u'KS', u'KY', u'LA', u'MA', u'MD', u'ME', u'MI', u'MN', u'MO', u'NE', u'NH', u'NJ', u'NV', u'NY', u'OH', u'OR', u'PA', u'SC', u'TN', u'TX', u'UT', u'VA', u'WA', u'WI', u'WV']
    len_states = len(states)
    for index, state in enumerate(states):
        flag = [0 for n in states]
        flag[index] = 1
        combine = request + flag
        state_matrix.update({state: int(logreg_production.predict(combine))})
    return state_matrix
```
